# 해시 테이블

## 입력 값을 해시 함수를 거쳐 정해진 값으로 변환하는 데이터 저장 형태
### 성능 좋은 해시 함수들의 특징

- 충돌의 최소화
- 쉽고 빠른 연산
- 해시 값들이 균일하게 분포
- 키 값의 모든 정보를 이용하여 해시
- 해시 테이블 사용 효율을 높일 것

# 대표적인 해시 함수 만드는 방법



## Load Factor

- 한 버킷 안에 몇 개의 키가 대응되는 지를 나타내는 지표
- 저장된 데이터 수 / 버킷의 수
- 너무 낮으면 버킷이 데이터에 비해 많아 메모리면에서 안좋음, 또한 하나의 버킷에 지나치게 많은 키가 대응될 가능성도 있다
- 너무 높을 경우 충돌이 자주 일어남을 의미.

```python
# 충돌이 일어나는 경우
# 칸이 365개인 생일의 경우도 23명만 있어도 같이 겹칠 확률이 50%가 된다. 

import random

trials=100000 # 실험횟수
same_birth=0

for _ in range(trials):
    births=[]
    # 실험당 23명의 사람이 생일이 한 번이라도 겹치는 지 확인
    for i in range(23):
        birth = random.randint(1,365)
        if birth in births:
            same_birth += 1
            break
        births.append(birth)
        
print(f'{same_birth/trials*100}%')
```
