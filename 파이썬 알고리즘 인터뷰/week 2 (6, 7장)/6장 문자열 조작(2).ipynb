```python
from collections import *
import re

paragraph = "Bob hit a ball, the hit BALL flew far after it was hit"
banned=['hit']
```

```python
# list comprehension과 re(정규표현식), Counter 함수에 대한 추가적인 공부 필요 

words = [word for word in re.sub(r'[^\w]',' ',paragraph).lower().split() if word not in banned] # 문자/숫자가 아닌 것들을 공백으로 대체
print(words)
Counter(words).most_common()[0]
```

```python
# 함수 형식으로 깔끔하게 정리 

def mostCommonWord(paragraph: str, banned) -> str:
    words = [word for word in re.sub(r'[^\w]',' ',paragraph).lower().split() if word not in banned] # 숫
    counts =collections.Counter(words)
    return counts.most_common(1)[0][0]

mostCommonWord(paragraph,banned)
```

```python
# list comprehension

# 기본 베이스 [변수를 통한 표현식, for 사용할 변수 , in 변수 범위, ]
evens = [x * 2 for x in range(11)]
print(evens)
print('-----------------------------------------------')

# if 조건문 추가 
from math import sqrt
non_squars = [x for x in range(101) if sqrt(x)**2 != x]
print(non_squars)
print('-----------------------------------------------')

# if 조건문 2개 넣기
# 조건문 사이에 and를 쓰지 않고 그대로 활용
arr = [n for n in range(1, 31) if n % 2 == 0 if n % 3 == 0]
print(arr)
print('-----------------------------------------------')

# or 조건문은 아예 활용이 안되서 한 if 조건문 안에 or 을 넣는다
arr = [n for n in range(1, 16) if n % 2 == 0 or n % 3 == 0]
# 한 if 문 내에서 or 연산 해결
print(arr)
print('-----------------------------------------------')

# else 구문 활용하기 
arr= [a if a else 2 for a in [0,1,0,3]]
print(arr)
print('-----------------------------------------------')

# for문 2개를 사용한 list comprehension
# 나올 수 있는 경우의 수를 모두 보여줌 
epithets = ['sweet', 'annoying', 'cool', 'grey-eyed']
names = ['john', 'alice', 'james']
epithet_names = [(e, n) for e in epithets for n in names]
print(epithet_names)
print('-----------------------------------------------')

# zip을 사용하면 순서대로 대응 될 때만 사용 됨. 
epithets = ['sweet', 'annoying', 'cool']
names = ['john', 'alice', 'james']
epithet_names = [(e, n) for e,n in zip(epithets, names)]
print(epithet_names)
print('-----------------------------------------------')

# 3개 활용하는 경우
# 피타고라스 방정식을 만족하는 경우
solutions = [(x, y, z) for x in range(1, 30) for y in range(x, 30) for z in range(y, 30) if x**2 + y**2 == z**2]
print(solutions)
```

```python
# set comprehension
# [] 대신 {} list 대신 set을 반환 

# 다음의 LC는 중복된 값들을 포함한다
no_primes = [j for i in range(2, 9) for j in range(i * 2, 50, i)]
print('list comprehension',no_primes)
print('-----------------------------------------------')
# SC를 사용하면 중복값이 없는 집합을 얻을 수 있다
no_primes = {j for i in range(2, 9) for j in range(i * 2, 50, i)}
print('set comprehension',no_primes)
```

```python
# dict comprehension
# set comprehension과 같이 {}을 활용하지만 표현식에서 key:value 방식을 활용

# 두 리스트를 하나의 dict로 합치는 DC. 하나는 key, 또 다른 하나는 value로 사용한다
subjects = ['math', 'history', 'english', 'computer engineering']
scores = [90, 80, 95, 100]
score_dict = {key: value for key, value in zip(subjects, scores)}
print(score_dict)

# 튜플 리스트를 dict 형태로 변환하는 DC
score_tuples = [('math', 90), ('history', 80), ('english', 95), ('computer engineering', 100)]
score_dict = {t[0]: t[1] for t in score_tuples}
print(score_dict)
```

# 6-5 그룹 애너그램

- 애너그램 : 문자를 재배열해서 다른 뜻을 가진 단어로 바꾸는 과정
- 풀이 방법 : 각 '문자를 알파벳 순서대로 정렬'해서 같은 단어끼리 묶는다.

```python
words = ['eat','tea','tan','ate','nat','bat']

def groupAnagrams(strs):
    anagrams = collections.defaultdict(list)         # default dict 생성
    
    for word in strs:
        anagrams[''.join(sorted(word))].append(word) # key에 정렬한 값, value에 해당 단어 리스트를 추가
    
    return anagrams.values(), anagrams.keys()

anagram_words, anagram_keys=groupAnagrams(words)
print(anagram_words)
print(anagram_keys)
```

```python
# sorted 함수 활용
# 특별한 경우가 아니면 파이썬 내장 sorted 함수를 사용하는 것이 정렬 속도에서 훨씬 더 빠르다

# 일반 숫자 정렬
a=[2,5,1,9,7]
print('숫자 정렬')
print(sorted(a))
print('-----------------------------------------------')

# 단어도 알파벳 순서대로 정렬이 가능하다
print('알파벳 정렬')
print(sorted('eat'))
print(sorted('ate'))
print('정렬후 문자로 결합')
print("".join(sorted('ate')))
print('-----------------------------------------------')

# key와 람다를 활용해 별도로 정렬 기준 설정 가능
c=['ccc','aaaa','d','bb']
print('일반적인 알파벳 순서')
print(sorted(c))
print('길이 기준으로 정렬')
print(sorted(c,key=len))
print('람다 활용 : 단어의 두 번째 알파벳 순서')
c1=['abaa','bcbb','cdcc','dadd']
print('디폴트 :', sorted(c1))
print('람다활용 두 번째 알파벳 순서 : ',sorted(c1,key=lambda x : x[1]))
print('key 두 가지 활용하기')
z=['cde','cfc','abc']
print('디폴트 : 앞에서 부터 알파벳 순',sorted(z))
print('순서 설정 :',sorted(z,key=lambda x : (x[0],x[-1])))
```

# 6-6

- 가장 긴 팰린드롬 부분 문자열
- 가장 긴 팰린드롬 부분 문자열을 출력하라
- 팰린드롬 : 앞으로 읽어도 반대로 읽어도 똑같은 단어

```python
# 최종 코드는 다음과 같다 

def longestPalindrome(s):
    # 팰린드롬 판별 및 투 포인터 확장
    def expand(left,right):
        while left >= 0 and right <=len(s) and s[left]==s[right-1]: # 양 사이드가 같으면 한 칸씩 넓히는 함수 
            left -=1 
            right +=1
        return s[left +1 : right-1]
    
    if len(s)<2 or s == s[::-1]:   # 입력된 문자의 길이가 1이거나 단어 전체가 팰린드롬인 경우를 미리 판단하여 속도 향상 
        return s                   # s[a:b:c] : a~b-1 인덱스 사이에 c 간격으로 추출, c가 음수면 반대 방향으로 
    
    result = ''
    
    for i in range(len(s)-1):
        result =max(result,expand(i,i+1),expand(i,i+2),key=len) # 왼쪽부터 순서대로 이동
                                                                # 기존 단어, 짝수 개수, 홀수 개수 중 가장 긴 단어(key=len)를 반환   
    return result 


longestPalindrome('323454322')
```

1. expand : 팰린드롬인지 아닌지를 먼저 판단후 팰린드롬이면, 왼/오른쪽으로 한 칸 씩 넓혀 해당 단어도 팰린드롬인지 확인하는 함수
2. 최종적으로 expand 함수는 지정된 중심 인덱스에서 가장 긴 팰린드롬을 반환
3. 이를 왼쪽부터 중심을 옮기면서 가장 긴 팰린드롬을 찾음

## [자주 사용하는 문자 클래스]
### 깊게 파고들게 아니면 이정도만 알아둬도 될듯

- [0-9] 또는 [a-zA-Z] 등은 무척 자주 사용하는 정규 표현식이다. 이렇게 자주 사용하는 정규식은 별도의 표기법으로 표현할 수 있다. 다음을 기억해 두자.

- \d - 숫자와 매치, [0-9]와 동일한 표현식이다.
- \D - 숫자가 아닌 것과 매치, [^0-9]와 동일한 표현식이다.
- \s - whitespace 문자와 매치, [ \t\n\r\f\v]와 동일한 표현식이다. 맨 앞의 빈 칸은 공백문자(space)를 의미한다.
- \S - whitespace 문자가 아닌 것과 매치, [^ \t\n\r\f\v]와 동일한 표현식이다.
- \w - 문자+숫자(alphanumeric)와 매치, [a-zA-Z0-9_]와 동일한 표현식이다.
- \W - 문자+숫자(alphanumeric)가 아닌 문자와 매치, [^a-zA-Z0-9_]와 동일한 표현식이다.

- 대문자로 사용된 것은 소문자의 반대임을 추측할 수 있다.

# Dot(.)
- 정규 표현식의 Dot(.) 메타 문자는 줄바꿈 문자인 \n을 제외한 모든 문자와 매치됨을 의미한다.
- a.b는 "a + 모든문자 + b" , a와 b라는 문자 사이에 어떤 문자가 들어가도 모두 매치된다는 의미
- ex) "aab","a0b"는 정규식과 매치 / "abc"는 "a"와 "b" 사이에 어떤 문자라도 하나는있어야 하는 이 정규식과 일치하지 않으므로 매치되지 않는다.
- 문자 클래스([]) 내에 Dot(.) 메타 문자가 사용된다면 이것은 "모든 문자"라는 의미가 아닌 문자 . 그대로를 의미
- ex) a[.]b => "a + Dot(.)문자 + b"

```python
p = re.compile('a.b')
li=["aab","a0b","abc"]
for i in li:
    m = p.match( i )
    if m:
        print(i, 'Match found: ', m.group())
    else:
        print(i, 'No match')
```

```python
p = re.compile('a[.]b')
li=["aab","a0b","abc",'a.b']
for i in li:
    m = p.match( i )
    if m:
        print(i, 'Match found: ', m.group())
    else:
        print(i, 'No match')
```

# 반복 (*)
- (*)바로 앞에 있는 문자가 0부터 무한대로 반복될 수 있다는 의미

```python
p = re.compile('ca*t')
li=['ct','cat','caaaat']
for i in li:
    m = p.match( i )
    if m:
        print(i, 'Match found: ', m.group())
    else:
        print(i, 'No match')
```

# 반복 (+)
- 최소 1번 이상 반복된다는 것을 의미

```python
p = re.compile('ca+t')
li=['ct','cat','caaaat']
for i in li:
    m = p.match( i )
    if m:
        print(i, 'Match found: ', m.group())
    else:
        print(i, 'No match')
```

# 반복 ({m,n})
- {}안에 숫자 하나만 있다면 해당 숫자만큼만 반복
- 두개 가 있다면 m 이상 n 이하만큼만 반복

```python
p = re.compile('ca{2}t')
li=['cat','caat','caaat']
for i in li:
    m = p.match( i )
    if m:
        print(i, 'Match found: ', m.group())
    else:
        print(i, 'No match')
```

```python
p = re.compile('ca{2,5}t')
li=['cat','caat','caaat']
for i in li:
    m = p.match( i )
    if m:
        print(i, 'Match found: ', m.group())
    else:
        print(i, 'No match')
```

```python
# 비슷한 개념으로 ?
# 없거나 1개 있다는 개념 ( 0 or 1)
p = re.compile('ca?t')
li=['ct','cat','caat']
for i in li:
    m = p.match( i )
    if m:
        print(i, 'Match found: ', m.group())
    else:
        print(i, 'No match')
```

# 자주 사용될만한 함수들
- match()	: 문자열의 처음부터 정규식과 매치되는지 조사한다.
- search()	: 문자열 전체를 검색하여 정규식과 매치되는지 조사한다.
- findall()	: 정규식과 매치되는 모든 문자열(substring)을 리스트로 돌려준다.
- sub() : 정규식과 매치되는 부분을 다른 문자로 쉽게 바꿀 수 있다.

```python
# match : 문자열 전체가 정규식과 매칭되는 지 조사
p = re.compile('[a-z]+')
m = p.match( 'string goes here' )
if m:
    print('Match found: ', m.group())
else:
    print('No match')
```

# 위의 함수와와 쓰이는 함수

- group() : 매치된 문자열을 돌려준다.
- start() : 매치된 문자열의 시작 위치를 돌려준다.
- end() : 매치된 문자열의 끝 위치를 돌려준다.
- span() : 매치된 문자열의 (시작, 끝)에 해당하는 튜플을 돌려준다.

```python
p = re.compile('[a-z]+')
m = p.match("python")
print(m.group())
print(m.start())
print(m.end())
print(m.span())
```

```python
p = re.compile('[a-z]+')
m = p.search("3python")
print(m.group())
print(m.start())
print(m.end())
print(m.span())
```
