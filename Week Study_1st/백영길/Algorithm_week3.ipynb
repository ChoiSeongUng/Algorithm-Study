```python
def Max_Heapify(A,i,n):
    """
    AÏùò ArrayÏóêÏÑú iÎ≤àÏß∏Î•º Ï†úÏô∏Ìïú ÏôºÏ™Ω, Ïò§Î•∏Ï™ΩÏùò SubtreeÍ∞Ä Heap Ï°∞Í±¥(Î∂ÄÎ™® ÎÖ∏ÎìúÍ∞Ä ÏûêÏãùÎ≥¥Îã§ ÌÅº)
    ÏùÑ ÎßåÏ°±Ìï† Îïå, Ï†ÑÏ≤¥ ArrayÍ∞Ä Heap Ï°∞Í±¥ÏùÑ ÎßåÏ°±ÌïòÎèÑÎ°ù ÌïòÎäî Ìï®Ïàò(nÎ≤àÏùò ÏúÑÏπòÍπåÏßÄ)
    """
    l = 2 * i + 1 # iÎ≤àÏß∏ indexÏóêÏÑú ÏôºÏ™Ω ÏûêÏãù ÏúÑÏπò(2i)
    r = 2 * i + 2 # iÎ≤àÏß∏ indexÏóêÏÑú Ïò§Î•∏Ï™Ω ÏûêÏãù ÏúÑÏπò(2i+1)
    if l < n and A[l] > A[i]:# ÏôºÏ™ΩÏùò ÏûêÏãùÎÖ∏ÎìúÍ∞Ä Î∂ÄÎ™®ÎÖ∏ÎìúÎ≥¥Îã§ ÌÅ¥ Îïå
        largest = l
    else:
        largest = i
    if r < n and A[r] > A[largest]:
        largest = r # Ïò§Î•∏Ï™ΩÏùò ÏûêÏãùÎÖ∏ÎìúÍ∞Ä Î∂ÄÎ™®ÎÖ∏Îìú/ÏôºÏ™ΩÎ≥¥Îã§ ÌÅ¥ Îïå 
    if largest != i:
        temp = A[i]
        A[i] = A[largest]
        A[largest] = temp
        Max_Heapify(A,largest,n)
```

```python
def Build_Max_Heap(A):
    n = len(A)
    for i in range(len(A)//2, -1, -1):
        Max_Heapify(A,i,n)
```

```python
a = [4,1,3,2,16,9,10,14,8,7]
Build_Max_Heap(a)
a
# Î∂ÄÎ™®ÎÖ∏ÎìúÍ∞Ä ÏûêÏãùÏùò ÎÖ∏ÎìúÎ≥¥Îã§ Ìï≠ÏÉÅ ÌÅ∞Í≤ÉÏùÑ ÎßåÏ°±ÌïòÍ≥† ÏûàÏùåÏùÑ Ïïå Ïàò ÏûàÎã§. 
```

```python
def Heap_sort(A):
    """
    Î≥µÏû°ÎèÑ O(nlog(n))Î•º Í∞ÄÏßÄÎäî Heap Ï†ïÎ†¨Íµ¨ÌòÑÌïòÎäî Ìï®Ïàò
    """
    Build_Max_Heap(A) # Î®ºÏ†Ä Heap Ï°∞Í±¥ÏùÑ ÎßåÏ°±ÌïòÎèÑÎ°ù ÏÑ§Ï†ï
    for i in range(len(A)-1,0,-1):
        temp = A[i]
        A[i] = A[0] 
        A[0] = temp 
        # Ï†§ ÌÅ∞ Ïàò (ÎøåÎ¶¨ÎÖ∏Îìú)Î•º Ï†§ Î∞ë(Ï†úÏùº ÏûëÏùÄ)Í∞íÍ≥º ÍµêÏ≤¥
        Max_Heapify(A,0,i) 
        # ÎøåÎ¶¨ÎÖ∏ÎìúÎ°ú Ïò¨ÎùºÏò® Ï†§ ÏûëÏùÄ Í∞íÏùÑ Îã§Ïãú ÏõêÏúÑÏπòÎ°ú Î≥¥ÎÉÑ.
```

```python
a = [4,1,3,2,16,9,10,14,8,7]
Heap_sort(a)
a
```

2. Max_HeapifyÏùÄ log(n)Ïùò Î≥µÏû°ÎèÑÎ•º Í∞ÄÏßÑÎã§. Ïù¥ÏßÑ Ìä∏Î¶¨ÏóêÏÑú Îç∞Ïù¥ÌÑ∞Î•º 2Î≤àÏî© ÎÇòÎàÑÍ≤å ÎêòÎØÄÎ°ú ÍπäÏù¥Îäî log(n)Ïù¥ ÎêòÍ≥† ÎøåÎ¶¨ ÍπåÏßÄ ÌÉêÏÉâÏùÑ Í±∞ÏπòÍ∏∞ ÎïåÎ¨∏Ïù¥Îã§.

### 1. Îçî ÎßµÍ≤å

Î≥∏ Î¨∏Ï†úÎäî Îã§ÏùåÍ≥º Í∞ôÎã§. 
1. Î™®Îì† ÏùåÏãùÏùò Ïä§ÏΩîÎπå ÏßÄÏàòÎ•º K Ïù¥ÏÉÅÏúºÎ°ú ÎßåÎì§Í≥†Ïûê ÌïúÎã§.
2. Î™®Îì† ÏùåÏãùÏùò Ïä§ÏΩîÎπå ÏßÄÏàòÎ•º K Ïù¥ÏÉÅÏúºÎ°ú ÎßåÎì§Í∏∞ ÏúÑÌï¥ Ïä§ÏΩîÎπå ÏßÄÏàòÍ∞Ä Í∞ÄÏû• ÎÇÆÏùÄ Îëê Í∞úÏùò ÏùåÏãùÏùÑ ÏÑûÎäîÎã§.
3. ÏÑûÏùÄ ÏùåÏãùÏùò Ïä§ÏΩîÎπå ÏßÄÏàò = Í∞ÄÏû• ÎßµÏßÄ ÏïäÏùÄ ÏùåÏãùÏùò Ïä§ÏΩîÎπå ÏßÄÏàò + (Îëê Î≤àÏß∏Î°ú ÎßµÏßÄ ÏïäÏùÄ ÏùåÏãùÏùò Ïä§ÏΩîÎπå ÏßÄÏàò * 2)
4. Î™®Îì† ÏùåÏãùÏùò Ïä§ÏΩîÎπå ÏßÄÏàòÍ∞Ä K Ïù¥ÏÉÅÏù¥ Îê† ÎïåÍπåÏßÄ Î∞òÎ≥µÌïòÏó¨ ÏÑûÎäîÎã§.
5. Ïù¥Î•º ÏúÑÌï¥ ÏÑûÏñ¥Ïïº ÌïòÎäî ÏµúÏÜå ÌöüÏàòÎ•º ReturnÌïúÎã§.
> Îã® Î™®Îì† ÏùåÏãùÏùò Ïä§ÏΩîÎπå ÏßÄÏàòÎ•º k Ïù¥ÏÉÅÏúºÎ°ú ÎßåÎì§ Ïàò ÏóÜÎäî Í≤ΩÏö∞ÏóêÎäî -1ÏùÑ return ÌïúÎã§.

### 1-1. ÎÇ¥ ÌíÄÏù¥ - Îç∞ÌÅ¨Î•º ÌôúÏö©
- Ï†ïÎãµÏùÄ ÎßûÏúºÎÇò, Ìö®Ïú®ÏÑ± ÌÖåÏä§Ìä∏ÏóêÏÑú ÏãúÍ∞Ñ Ï¥àÍ≥ºÍ∞Ä Î∞úÏÉù

Ïö∞ÏÑ†ÏàúÏúÑ Ìûô Íµ¨Ï°∞Í∞Ä ÏïÑÎãàÎùºÏÑú, **Îß§Î≤à Ï†ïÎ†¨ÌïòÍ≥† Îã§Ïãú Îç∞ÌÅ¨ Íµ¨Ï°∞Î°ú Î≥ÄÌôòÌïòÎäî Í≥ºÏ†ïÏóêÏÑú Ìö®Ïú®ÏÑ±Ïù¥ Îñ®Ïñ¥Ïßê** 


```python
from collections import deque
def solution(scoville, K):
    answer = 0
    scoville = deque(sorted(scoville))
    # Ï†ïÎ†¨ÌïòÍ≥† Îç∞ÌÅ¨ Íµ¨Ï°∞Î°ú Î≥ÄÌôò
    while True:
        least = scoville.popleft()
        if least >= K:
            break
        # Í∞ÄÏû• ÏûëÏùÄ ÏõêÏÜåÍ∞Ä KÎ≥¥Îã§ ÌÅ∞ Í≤ΩÏö∞ Ï¢ÖÎ£å
        if len(scoville) == 0:
            return -1
        # Í∞ÄÏû• ÏûëÏùÄ ÏõêÏÜåÍ∞Ä KÎ≥¥Îã§ ÌÅ¨Í≤å ÏÑûÏßÄ Î™ªÌïòÎäî Í≤ΩÏö∞ -1
        less = scoville.popleft()
        scoville.appendleft(least + 2*less)
        scoville = deque(sorted(scoville))
        # Îã§Ïãú Ï†ïÎ†¨ÌïòÍ≥† Îç∞ÌÅ¨Î°ú ÎßåÎì¶
        answer += 1
    return answer
```

```python
# Test Data 
solution([1, 2, 3, 9, 10, 12],7)
```

```python
import heapq
def solution(scoville, K):
    answer = 0
    heapq.heapify(scoville)
    # Ï†ïÎ†¨ÌïòÍ≥† Îç∞ÌÅ¨ Íµ¨Ï°∞Î°ú Î≥ÄÌôò
    while True:
        least = heapq.heappop(scoville)
        if least >= K:
            break
        # Í∞ÄÏû• ÏûëÏùÄ ÏõêÏÜåÍ∞Ä KÎ≥¥Îã§ ÌÅ∞ Í≤ΩÏö∞ Ï¢ÖÎ£å
        if len(scoville) == 0:
            return -1
        # Í∞ÄÏû• ÏûëÏùÄ ÏõêÏÜåÍ∞Ä KÎ≥¥Îã§ ÌÅ¨Í≤å ÏÑûÏßÄ Î™ªÌïòÎäî Í≤ΩÏö∞ -1
        less = heapq.heappop(scoville)
        heapq.heappush(scoville, least + 2*less)
        answer += 1
    return answer
```

```python
def solution(jobs):
    jobs = sorted(jobs, key = lambda x: x[1])
    time = 0; answer = 0
    for i in range(len(jobs)):
        if jobs[i][0] <= time:
            time += jobs[i][1]
            answer += time - jobs[i][0]
        else:
            time += 1
    return answer//len(jobs)
```

```python
a = [[1, 10], [3, 3], [10, 3]]
solution(a)
```
